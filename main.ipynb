# Buscar com Eurísticas

```python
# Criando vértices
A = Euristica("A", 5)
B = Euristica("B", 3)
C = Euristica("C", 2)
D = Euristica("D", 0)

# Definindo as arestas
A.insere_adjacente_ordenado(B, 2)
A.insere_adjacente_ordenado(C, 4)
B.insere_adjacente_ordenado(D, 1)
C.insere_adjacente_ordenado(D, 5)
```

```python
# Realizando a busca gulosa
A.busca_gulosa(D)
```
